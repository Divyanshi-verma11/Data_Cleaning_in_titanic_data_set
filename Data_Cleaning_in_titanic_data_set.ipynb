# normalization technique

# min max algorithm

```python
data = pd.read_csv("titanic.csv")
new_fare = (data["fare"] - np.min(data["fare"])) / (np.max(data["fare"])-np.min(data["fare"]))
data["new_fare"] = new_fare
new_fare
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["fare"])

plt.subplot(1,2,1)
plt.title("before")
sns.distplot(data["new_fare"])
plt.show()
```

# using scikit_learn library

```python
from sklearn.preprocessing import MinMaxScaler
n = MinMaxScaler()
n.fit(data[["fare"]])
new_fare = n.transform(data[["fare"]])
data["new_fare"] = new_fare
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(data["fare"])

plt.subplot(1,2,1)
plt.title("before")
sns.distplot(data["new_fare"])
plt.show()
```
